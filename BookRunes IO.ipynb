```python
#Setup E-mail Log

import pandas as pd
from matplotlib import pyplot as plt
logs = pd.read_csv("Logs.csv")

logs["date"] = pd.to_datetime(logs.date)
logs["day"] = logs.date.dt.day
```

```python
#Get logs for all days

june_26 = logs[logs.day == 26]
june_25 = logs[logs.day == 25]
june_24 = logs[logs.day == 24]
june_23 = logs[logs.day == 23]
june_22 = logs[logs.day == 22]
june_21 = logs[logs.day == 21]
june_20 = logs[logs.day == 20]

june_26_opens = june_26[june_26.status == "Opened"]
june_25_opens = june_25[june_25.status == "Opened"]
june_24_opens = june_24[june_24.status == "Opened"]
june_23_opens = june_23[june_23.status == "Opened"]
june_22_opens = june_22[june_22.status == "Opened"]
june_21_opens = june_21[june_21.status == "Opened"]
june_20_opens = june_20[june_20.status == "Opened"]



days_data = [june_20, june_21, june_22, june_23, june_24, june_25, june_26]
days_opens = [june_20_opens, june_21_opens, june_22_opens, june_23_opens, june_24_opens, june_25_opens, june_26_opens]
```

```python
#Graph lists

dates = ["06/20", "06/21", "06/22", "06/23", "06/24", "06/25", "06/26"]
dates2 = ["06/21", "06/22", "06/23", "06/24", "06/25"]

total_emails = []
for day in days_data:
    total_emails.append(len(day))

total_opens = []
for day in days_opens:
    total_opens.append(len(day))
```

```python
with plt.style.context("Solarize_Light2"): 
    plt.plot(dates, total_emails)
    plt.title("E-mails Submitted")
    plt.xlabel("Date")
    plt.ylabel("Total E-mails")
    plt.savefig("bookrunes_analytics/static/total_emails_submitted", dpi=1200, edgecolor="#002b36", facecolor="#fdf6e3")
```

```python
#Totals for Specific Providers

june_26_gmail = june_26[june_26.to.str.contains("gmail")]
june_25_gmail = june_25[june_25.to.str.contains("gmail")]
june_24_gmail = june_24[june_24.to.str.contains("gmail")]
june_23_gmail = june_23[june_23.to.str.contains("gmail")]
june_22_gmail = june_22[june_22.to.str.contains("gmail")]
june_21_gmail = june_21[june_21.to.str.contains("gmail")]
june_20_gmail = june_20[june_20.to.str.contains("gmail")]

june_26_yahoo = june_26[june_26.to.str.contains("yahoo")]
june_25_yahoo = june_25[june_25.to.str.contains("yahoo")]
june_24_yahoo = june_24[june_24.to.str.contains("yahoo")]
june_23_yahoo = june_23[june_23.to.str.contains("yahoo")]
june_22_yahoo = june_22[june_22.to.str.contains("yahoo")]
june_21_yahoo = june_21[june_21.to.str.contains("yahoo")]
june_20_yahoo = june_20[june_20.to.str.contains("yahoo")]

june_26_outlook = june_26[june_26.to.str.contains("outlook")]
june_25_outlook = june_25[june_25.to.str.contains("outlook")]
june_24_outlook = june_24[june_24.to.str.contains("outlook")]
june_23_outlook = june_23[june_23.to.str.contains("outlook")]
june_22_outlook = june_22[june_22.to.str.contains("outlook")]
june_21_outlook = june_21[june_21.to.str.contains("outlook")]
june_20_outlook = june_20[june_20.to.str.contains("outlook")]

gmail_submitted = [june_20_gmail, june_21_gmail, june_22_gmail, june_23_gmail, june_24_gmail, june_25_gmail, june_26_gmail]
gmail_totals = []

for day in gmail_submitted:
    gmail_totals.append((len(day) / 6103 * 100) - 100)
    
yahoo_submitted = [june_20_yahoo, june_21_yahoo, june_22_yahoo, june_23_yahoo, june_24_yahoo, june_25_yahoo, june_26_yahoo]
yahoo_totals = []

for day in yahoo_submitted:
    yahoo_totals.append((len(day) / 2884 * 100) - 100)
    
outlook_submitted = [june_20_outlook, june_21_outlook, june_22_outlook, june_23_outlook, june_24_outlook, june_25_outlook, june_26_outlook]
outlook_totals = []

for day in outlook_submitted:
    outlook_totals.append((len(day) / 315 * 100) - 100)
```
